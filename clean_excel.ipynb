```python
import pandas as pd
import re
```

```python
df = pd.read_csv('data_for_clean_excel.csv')
df
```

```python
df = df.dropna()
```

```python
df.columns = df.columns.str.strip().str.title()
df
```

```python
df = df.applymap(lambda x:x.strip() if isinstance(x,str) else x)
df
```

```python
df['Customer Name'] = df['Customer Name'].str.title()
df['City'] = df['City'].str.title()
df
```

```python
def clean_email(email):
    if pd.isna(email):
        return "Invalid"
    email = email.lower()
    pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
    return email if re.match(pattern,email) else "Invalid"

df["Email"] = df["Email"].apply(clean_email)
```

```python
def clean_phone(phone):
    if pd.isna(phone):
        return "Invalid"
    digits = re.sub(r"\D", "", phone)
    if digits.startswith("92"):
        digits="0"+digits[2:]
    if len(digits)==11 and digits.startswith("03"):
        return digits
    else:
        return "Invalid"

df["Phone"] = df["Phone"].apply(clean_phone)
```

```python
df["Amount"] = pd.to_numeric(df["Amount"], errors='coerce').fillna(0).astype('int')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce',)
df
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.title()
)
```

```python
merged = (
    df.groupby("Customer Name", as_index=False)
    .agg({
        "Customer Name": "first",
        "Phone": "first",
        "City": "last",
        "Amount": "sum",
        "Date": "max"
    })
)

merged = merged.sort_values("Amount", ascending=False)
merged
```

```python
column_order = [
    "Customer Name",
    "Email",
    "Phone",
    "City",
    "Amount",
    "Date"
]

merged = merged.reindex(columns=column_order)
merged
```

```python
merged.to_csv("Final File.csv")
```
